### Cleaning Data

####Pengisian data-data kosong pada dataset yang diberikan

Pada tahap ini, akan dilakukan pengisian dataset yang masih kosong. Untuk parameter "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES" ,"DEWP" ,"RAIN" , "WSPM", "PM2.5" akan diisi oleh mean dari masing-masing data. Untuk "wd" akan diisi menggunakan modus dari data dikarenakan data merupakan tipe object

```python
#pengisian data-data yang kosong
parameters =["PM10","SO2","NO2","CO","O3","TEMP","PRES","DEWP","RAIN", "WSPM", "PM2.5"]
def pengisian_data_kosong(nama):
  for parameter in parameters:
    nama[parameter].fillna(nama[parameter].mean(), inplace=True) #Pengisian data kosong pada parameter menggunakan mean
  nama.wd.fillna(value= nama.wd.mode()[0], inplace=True) #pengisian data kosong pada wd dengan modus
  #Pengubahan parameter pada date_times diubah ke type datetime
  nama["datetime"] = pd.to_datetime(nama[["year", "month", "day", "hour"]])
```

### Pertanyaan 2:

```python
#mendefinisikan fungsi untuk perhitungan setiap polutan di stasiun
def rata_rata(nama):
  senyawa = ('PM2.5', 'PM10', 'SO2', 'NO2', 'CO','O3')
  rata_rata_senyawa = (nama['PM2.5'].mean(), nama['PM10'].mean(), nama['SO2'].mean(),
                 nama['NO2'].mean(), nama['CO'].mean(), nama['O3'].mean())
  return senyawa, rata_rata_senyawa
```
